```python
# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import cv2 as cv
import pandas as pd
import os
```

```python
df_train = pd.read_csv("training.csv", index_col=0)
```

```python
df_train.head()
```

```python
i =0
data_train = []
for img_name in df_train.index:
    img  = cv.imread("train_img/"+img_name,1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#     print(img.shape)
    data_train.append(img)
    if i == 100:
        break
    i +=1
data_t = np.asarray(data_train)
```

```python
data_t.shape
```

```python
plt.figure(figsize=(16,20)) # width,height : (20,10) 
for i in range(20):
    plt.subplot(5,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(data_t[i], cmap=plt.cm.binary)
    plt.grid('off')
```
